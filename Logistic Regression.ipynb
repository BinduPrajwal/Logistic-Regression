```python
bank1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome'])
bank1
```

```python
bank1['default'] = np.where(bank1['default'].str.contains("yes"), 1, 0)
bank1['housing'] = np.where(bank1['housing'].str.contains("yes"), 1, 0)
bank1['loan'] = np.where(bank1['loan'].str.contains("yes"), 1, 0)
bank1['y'] = np.where(bank1['y'].str.contains("yes"), 1, 0)
bank1
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
bank1=bank1.replace(order)
bank1
```

```python
bank1.info()
```

# ModelBuilding

```python
features=pd.concat([bank1.iloc[:,0:11],bank1.iloc[:,12:]],axis=1)
target=bank1.iloc[:,11]
```

```python
target
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model=LogisticRegression()
model.fit(features,target)
```

```python
pred=model.predict(features)
```

```python
pred
```

# Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target,pred))
```
